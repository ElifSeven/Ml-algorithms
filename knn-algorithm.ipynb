```python
from sklearn.model_selection import train_test_split
```

```python
X,y = mglearn.datasets.make_forge()
X_egitim,X_test,y_egitim,y_test = train_test_split(X,y,random_state=0)
from sklearn.neighbors import KNeighborsClassifier
snf = KNeighborsClassifier(n_neighbors=3)
snf.fit(X_egitim,y_egitim)
```

```python
snf.predict(X_test) # test verisindeki her bir nokta için en yakın komşuluklar bulundu
```

```python
snf.score(X_test,y_test) # yeni bir verinin sınıfını tahmin ederken kurduğumuz model %86 oranında doğru tahmin edeceğini gösteriyor.
```
